```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols

# Example 1: T-Test and Regression
# --------------------------------

# Create the dataframe
data = {
    'Type': ['Control', 'Control', 'Control', 'Control', 'Mutant', 'Mutant', 'Mutant', 'Mutant'],
    'Weight': [2.4, 3.5, 4.4, 4.9, 1.7, 2.8, 3.2, 3.9],
    'Size': [1.9, 3, 2.9, 3.7, 2.8, 3.3, 3.9, 4.8]
}
df = pd.DataFrame(data)
df
```

```python
# Perform t-test and regression using formula API
formula = 'Size ~ Type + Weight'
model = ols(formula, data=df).fit()
print(model.summary())
```

```python
# Create the dataframe
data = {
    'Lab': ['A', 'A', 'A', 'A', 'A', 'A', 'B', 'B', 'B', 'B', 'B', 'B'],
    'Type': ['Control', 'Mutant', 'Control', 'Mutant', 'Control', 'Mutant',
             'Control', 'Mutant', 'Control', 'Mutant', 'Control', 'Mutant'],
    'Expression': [1.7, 2, 2.2, 3.1, 3.6, 3.9, 0.9, 1.2, 1.9, 1.8, 2.2, 2.9]
}
df = pd.DataFrame(data)
df
```

```python
# Perform multiple linear regression using formula API
formula = 'Expression ~ Lab + Type'
model = ols(formula, data=df).fit()
print(model.summary())
```
